From typical user's perspective, the `lunapi` (here aliased `lp`) Python package comprises of three primary entities:

1. __A small set of utility functions:__  Most importantly, to initiate a new project (`lp.proc()`) alongside some other utility functions.

2. ___Project-level_ class:__ This class (`proj`) manages the creation of instance-level classes (i.e. objects reflecting individual EDFs), as well as sample lists.  Other member functions apply Luna commands to a sample list, or set project-level options.

3. ___Instance-level_ class:__ This class (`inst`) points to an individual, i.e. an EDF/annotation dataset.  

## High- and low-level moduldes

The `lunapi` package consists of two modules: a high-level interface (`lunapi1`) and a lower-level set of functions that directly interact with the C/C++ API (`lunapi0`).  

We expect most uses of this package will involve the higher set of `lunapi` functions.  The `lunapi` module is a pure Python module, which mainly contains wrappers around the lower-level calls, for example formatting outputs as Pandas dataframes rather than returning sets of base Python objects, e.g. lists of lists, etc.  There may be cases where using the lower-level `lunapi0` is more convenient, and so these are also exposed to package users.  All `lunapi0` functions are generated by the [pybind11]() package, as bindings to the C/C++ Luna API.   

The two primary classes described above (`proj` and `inst`) are defined in both the high- and low-level modules. 

In terms of package source, `lunapi1` is implemented in the Python file 
[lunapi/lunapi1.py](https://github.com/remnrem/luna-api/blob/c592e23fe3ecec03774ed245062cab797aee200c/src/lunapi/lunapi1.py)
whereas `lunapi0` is imeplemted in the C/C++ file
[lunapi/lunapi0.cpp](https://github.com/remnrem/luna-api/blob/c592e23fe3ecec03774ed245062cab797aee200c/src/lunapi/lunapi0.cpp)

Most users can ignore the distinction between so-called _level-1_ and _level-0_ functions, but they are described here for completeness.
